# Basic SymPlot test page

Here we set up our really preliminary test page for SymPlot.

Supported types:

* Triangle
* Polygon and RegularPolygon
* SymPy Expressions in one variable (x)


## Plot some regular polygons

```python
import symplot
from sympy import Point, sqrt, pi
from sympy.geometry import Triangle, RegularPolygon
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
t = Triangle(Point(0,0), Point(10, 0), Point(5, 5 * sqrt(3)))
ax = symplot.plot(t, color="blue")
pentagon = RegularPolygon(Point(5,2.6), 3, 5)
pentagon.spin(pi/10)
ax = symplot.plot(pentagon, ax, color="black", linewidth=.8)
```

## Plot a Circle

```python
from sympy.geometry import Circle
from sympy import symbols
import numpy as np

x = symbols('x')
x = 5
big_circle = Circle(Point(-20,+30), x * 2)
_ = symplot.plot(big_circle)
```

## Irregular Polygons

```python
# An irregular Polygon
from sympy.geometry import Polygon
v1 = Point(2,3)
v2 = Point(-3, 1)
v3 = Point(4,1)
v4 = Point(0, -3)
t = Polygon(v1, v2, v3, v4)
symplot.plot(t, color="gray")
```

# SymPy Expressions
Assuming they are expressions in one variable, x, We can also plot SymPy Expressions.  The x and y range that's visible will currently be in the closed interval, [-10, 10].  More needs to be done to make this settable.

```python
from sympy import symbols
x = symbols('x')
f = x**3 - x/2
_ = symplot.plot(f)
```
